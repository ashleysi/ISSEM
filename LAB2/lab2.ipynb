# Q1.d 
What is socket library in Python? Why is it used? Give one brief example.

The socket library in Python facilitates low-level network communication using the TCP/IP protocol. It serves as the foundation for building networking applications in Python, allowing for the creation of both servers and clients.

Usage:
Communication: Provides mechanisms for machines to communicate over networks.
Protocols: Enables developers to implement or customize networking protocols.


```python
import socket
import threading

def server():
    s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    s.bind(('localhost', 12347))
    s.listen(1)
    c, addr = s.accept()
    data = c.recv(1024)
    print("Server received:", data.decode())
    c.sendall(b'connected')
    c.close()

def client():
    s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    s.connect(('localhost', 12347))
    s.sendall(b'Hello--!')
    data = s.recv(1024)
    print("Client received:", data.decode())
    s.close()

# Running server and client concurrently in the notebook
threading.Thread(target=server).start()
threading.Thread(target=client).start()
```

# Q1.e
What is multithreading? 
Using Threading library, write a Python program to calculate the factorial of number 21 and 
square root of number 91.

Multithreading:
Multithreading is a technique in which multiple threads in a process execute tasks concurrently. A thread is the smallest unit of CPU execution. While multithreaded applications share the same memory space, each thread has its own set of local variables, program counter, and call stack. Multithreading can lead to better utilization of CPU resources, especially when tasks involve waiting, such as I/O operations. However, care must be taken to manage race conditions, deadlocks, and other concurrency-related issues.

```python
import threading
import math

def factorial(num):
    result = 1
    for i in range(1, num + 1):
        result *= i
    print(f"Factorial of {num} is {result}")

def square_root(num):
    result = math.sqrt(num)
    print(f"Square root of {num} is {result:.2f}")

# Create threads for the factorial and square root calculations
thread1 = threading.Thread(target=factorial, args=(21,))
thread2 = threading.Thread(target=square_root, args=(91,))

# Start the threads
thread1.start()
thread2.start()

# Wait for both threads to complete
thread1.join()
thread2.join()

print("Both tasks completed.")
```

## Differences between Symmetric and Asymmetric Encryption:

### Key Usage:

- **Symmetric:** Uses a single key for both encryption and decryption.
- **Asymmetric:** Utilizes a public key for encryption and a private key for decryption.

### Speed:

- **Symmetric:** Generally faster in terms of encryption and decryption.
- **Asymmetric:** Slower due to the complexity of the algorithms involved.

### Key Distribution:

- **Symmetric:** Presents a challenge in securely sharing the key since both parties need to possess the same key.
- **Asymmetric:** Allows the public key to be openly shared, simplifying the distribution process.

### Preference:

Asymmetric encryption offers a significant advantage in key distribution, making it more suitable for secure communication between unfamiliar parties. However, the choice between symmetric and asymmetric encryption ultimately depends on the specific use case, and many systems combine both types to achieve optimized security.

"""
Basic example showing how to read and validate data from a file using Pydantic.
"""

import json
from typing import List, Optional
import pydantic

# Custom exception for handling scenarios where both ISBN-10 and ISBN-13 are missing.
class ISBNMissingError(Exception):
    def __init__(self, title: str, message: str) -> None:
        self.title = title
        self.message = message
        super().__init__(message)

# Custom exception for handling incorrect ISBN-10 format.
class ISBN10FormatError(Exception):
    def __init__(self, value: str, message: str) -> None:
        self.value = value
        self.message = message
        super().__init__(message)

# Pydantic model for Author data.
class Author(pydantic.BaseModel):
    name: str
    verified: bool

# Pydantic model for Book data.
class Book(pydantic.BaseModel):
    title: str
    author: str
    publisher: str
    price: float
    # Optional fields.
    isbn_10: Optional[str]
    isbn_13: Optional[str]
    subtitle: Optional[str]
    author2: Optional[Author]

    # Validator to ensure either isbn_10 or isbn_13 is present.
    @pydantic.root_validator(pre=True)
    @classmethod
    def check_isbn_10_or_13(cls, values):
        if "isbn_10" not in values and "isbn_13" not in values:
            raise ISBNMissingError(
                title=values["title"],
                message="Document should have either an ISBN10 or ISBN13",
            )
        return values

    # Validator for ISBN-10 format and its properties.
    @pydantic.validator("isbn_10")
    @classmethod
    def isbn_10_valid(cls, value) -> None:
        chars = [c for c in value if c in "0123456789Xx"]
        if len(chars) != 10:
            raise ISBN10FormatError(value=value, message="ISBN10 should be 10 digits.")

        def char_to_int(char: str) -> int:
            if char in "Xx":
                return 10
            return int(char)

        if sum((10 - i) * char_to_int(x) for i, x in enumerate(chars)) % 11 != 0:
            raise ISBN10FormatError(
                value=value, message="ISBN10 digit sum should be divisible by 11."
            )
        return value

    # Configuration for this Pydantic model.
    class Config:
        allow_mutation = False  # Instances of this model are immutable.
        anystr_lower = True     # Convert any string to lowercase.

# Function to read and validate book data from a JSON file.
def main() -> None:
    with open("./data.json") as file:
        data = json.load(file)
        # Convert and validate data using the Book model.
        books: List[Book] = [Book(**item) for item in data]
        # Sample outputs.
        print(books)
        print(books[0])
        print(books[0].dict(exclude={"price"}))
        print(books[1].copy())

if __name__ == "__main__":
    main()

### Overview:

The code demonstrates the usage of `pydantic` to define and validate data models. The primary use case is reading and validating book data from a JSON file.

### Classes and their purpose:

1. **ISBNMissingError & ISBN10FormatError**: 
   - These are custom exception classes to handle specific error scenarios related to ISBN values in the data.
  
2. **Author**:
   - A `pydantic` model representing an author.
   - Fields: 
     - `name`: Name of the author.
     - `verified`: A boolean to indicate if the author's details have been verified.
  
3. **Book**:
   - A `pydantic` model representing a book.
   - Fields include `title`, `author`, `publisher`, `price`, and optional fields like `isbn_10`, `isbn_13`, `subtitle`, and `author2`.
   - Validators:
     - `check_isbn_10_or_13`: A root validator ensuring that either `isbn_10` or `isbn_13` is present in the data.
     - `isbn_10_valid`: Validates the `isbn_10` field format and checks if its sum is divisible by 11, a basic ISBN-10 verification step.
   - Config:
     - This inner class provides configuration settings for the `Book` model. 
     - `allow_mutation`: Prevents modification of instances once created.
     - `anystr_lower`: Converts any string field to lowercase.

4. **main() function**:
   - Reads book data from a file named "data.json".
   - Validates the data using the `Book` model.
   - Demonstrates some `pydantic` functionalities like converting models back to dictionaries and copying model instances.

### How Pydantic is used:

- **Data Modeling**: `pydantic` is primarily used to define data models (`Author` and `Book`) that dictate the structure and types of data.
  
- **Validation**: Custom validators (`check_isbn_10_or_13` and `isbn_10_valid`) are used to perform more nuanced validation that can't be expressed with type annotations alone.

- **Data Transformation**: The data read from the JSON file is transformed into `pydantic` model instances, allowing for type checks, validation, and easy data manipulation.

- **Configuration**: The `Config` inner class in the `Book` model shows how you can configure certain behaviors of `pydantic` models.

### Conclusion:

The code is a demonstration of how `pydantic` can help in ensuring data correctness and consistency by validating data against well-defined models. It effectively showcases how to handle complex data structures, validate them, and manage potential errors that can arise from data discrepancies.
